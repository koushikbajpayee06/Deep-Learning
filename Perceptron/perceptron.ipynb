```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 1000

cgpa = np.round(np.random.uniform(5.0, 10.0, n), 2)
resume_score = np.random.randint(40, 101, n)

# Linear decision boundary
placed = ((10 * cgpa + resume_score) >= 150).astype(int)

df = pd.DataFrame({
    "cgpa": cgpa,
    "resume_score": resume_score,
    "placed": placed
})

df.to_csv("placement_data.csv", index=False)

print(df.head())
```

```python
df = pd.read_csv("placement_data.csv")
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    data=df,
    x='cgpa',
    y='resume_score',
    hue='placed'
)

plt.show()
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
plt.xlabel('CGPA')
plt.ylabel('Resume Score')
plt.title('Perceptron Decision Boundary')
plt.show()
```
