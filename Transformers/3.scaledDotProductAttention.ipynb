```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
    # Ensure numerical stability by substracting the max value from each element
    e_x = np.exp(x - np.max(x))
    softmax_values = e_x / e_x.sum(axis=0) # Normalize to get probabilities
    percent_values = softmax_values*100 # Convert to percentage
    # Format the percentage values to two decimal places followed by a percentage sign
    formatted_parcent_values = [f"{value:.2f}%" for value in percent_values]
    return formatted_parcent_values
```

```python
softmax(np.array([4,5]))
```

```python
softmax(np.array([1,10]))
```
