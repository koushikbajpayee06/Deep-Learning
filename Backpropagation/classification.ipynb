```python
import pandas as pd
import numpy as np
```

```python
# Binary Classification Dataset: Placement
placement_df = pd.DataFrame([
    [8.5, 9.0, 1],
    [7.0, 7.5, 1],
    [6.0, 6.0, 0],
    [5.5, 5.0, 0]
], columns=['cgpa', 'profile_score', 'placement'])

placement_df
```

```python
# Sigmoid activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Binary Cross Entropy loss
def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# Initialize parameters
def initialize_parameters(input_dim, hidden_dim, output_dim):
    np.random.seed(42)
    parameters = {
        'W1': np.random.randn(hidden_dim, input_dim) * 0.01,
        'b1': np.zeros((hidden_dim, 1)),
        'W2': np.random.randn(output_dim, hidden_dim) * 0.01,
        'b2': np.zeros((output_dim, 1))
    }
    return parameters

# Forward propagation
def forward_propagation(X, parameters):
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    
    # Layer 1
    Z1 = np.dot(W1, X) + b1
    A1 = sigmoid(Z1)
    
    # Layer 2
    Z2 = np.dot(W2, A1) + b2
    A2 = sigmoid(Z2)
    
    cache = {'Z1': Z1, 'A1': A1, 'Z2': Z2, 'A2': A2, 'X': X}
    return A2, cache

# Backward propagation
def backward_propagation(y, cache, parameters):
    m = y.shape[1]
    W2 = parameters['W2']
    
    A1 = cache['A1']
    A2 = cache['A2']
    X = cache['X']
    Z1 = cache['Z1']
    
    # Output layer gradient
    dZ2 = A2 - y
    dW2 = np.dot(dZ2, A1.T) / m
    db2 = np.sum(dZ2, axis=1, keepdims=True) / m
    
    # Hidden layer gradient
    dA1 = np.dot(W2.T, dZ2)
    dZ1 = dA1 * A1 * (1 - A1)
    dW1 = np.dot(dZ1, X.T) / m
    db1 = np.sum(dZ1, axis=1, keepdims=True) / m
    
    grads = {'dW1': dW1, 'db1': db1, 'dW2': dW2, 'db2': db2}
    return grads

# Update parameters
def update_parameters(parameters, grads, learning_rate):
    parameters['W1'] -= learning_rate * grads['dW1']
    parameters['b1'] -= learning_rate * grads['db1']
    parameters['W2'] -= learning_rate * grads['dW2']
    parameters['b2'] -= learning_rate * grads['db2']
    return parameters

print("Manual neural network functions defined successfully!")
```

```python
# Prepare data
X = placement_df[['cgpa', 'profile_score']].values.T
y = placement_df[['placement']].values.T

# Initialize network
parameters = initialize_parameters(input_dim=2, hidden_dim=4, output_dim=1)

# Training loop
learning_rate = 0.5
epochs = 100
losses = []

for epoch in range(epochs):
    # Forward propagation
    y_pred, cache = forward_propagation(X, parameters)
    
    # Compute loss
    loss = binary_cross_entropy(y, y_pred)
    losses.append(loss)
    
    # Backward propagation
    grads = backward_propagation(y, cache, parameters)
    
    # Update parameters
    parameters = update_parameters(parameters, grads, learning_rate)
    
    if (epoch + 1) % 20 == 0:
        print(f"Epoch {epoch + 1}/{epochs}, Loss: {loss:.6f}")

print("\nTraining completed!")
print(f"Final Loss: {losses[-1]:.6f}")
```

```python
# Test predictions
y_pred, _ = forward_propagation(X, parameters)
y_pred_class = (y_pred > 0.5).astype(int)

# Display results
results = pd.DataFrame({
    'CGPA': X[0],
    'Profile Score': X[1],
    'Actual': y[0],
    'Predicted': y_pred[0].round(4),
    'Predicted Class': y_pred_class[0]
})
print(results)
```
