```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa','profile_score','lpa'])
```

```python
df
```

```python
def initilize_parameters(layer_dims):
    np.random.seed(3)
    parameters = {}
    L = len(layer_dims)
    for l in range(1, L):
        parameters['W' + str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1]) * 0.01
        parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))
    return parameters
```

```python
initilize_parameters([2,2,1])
```

```python
def linear_forward(A_prev, W, b):
    Z = np.dot(W, A_prev) + b
    return Z
```

```python
def forward_propagation(X, parameters):
    A = X
    L = len(parameters) // 2
    for l in range(1, L + 1):
        A_prev = A
        W = parameters['W' + str(l)]
        b = parameters['b' + str(l)]
        # print("A" + str(l-1) + " = ", A_prev)
        # print("W" + str(l) + " = ", W)
        # print("b" + str(l) + " = ", b)
        # print("--" * 20)
        A = linear_forward(A_prev, W, b)
        # print("A" + str(l) + " = ", A)
        # print("--" * 20)
    return A, A_prev
```

```python
X = df[['cgpa','profile_score']].values[0].reshape(2,1)
y = df[['lpa']].values[0][0]
parameters = initilize_parameters([2,2,1])
y_hat, A1=forward_propagation(X,parameters)
```

```python
y_hat = y_hat[0][0]
```

```python
def update_parameters(parameters, y, y_hat, A1, X, learning_rate=0.001):
    dZ2 = y_hat - y
    dW2 = np.dot(dZ2, A1.T)
    db2 = dZ2
    dZ1 = np.dot(parameters['W2'].T, dZ2)
    dW1 = np.dot(dZ1, X.T)
    db1 = dZ1
    parameters['W2'] -= learning_rate * dW2
    parameters['b2'] -= learning_rate * db2
    parameters['W1'] -= learning_rate * dW1
    parameters['b1'] -= learning_rate * db1
    return parameters
```

```python
update_parameters(parameters,y,y_hat,A1,X)
```

```python
# epochs implementation
parameters = initilize_parameters([2,2,1])
epochs = 5
for i in range(epochs):
    Loss = []
    for j in range(df.shape[0]):
        X = df[['cgpa','profile_score']].values[j].reshape(2,1)
        y = df[['lpa']].values[j][0]
        y_hat, A1=forward_propagation(X,parameters)
        y_hat = y_hat[0][0]
        Loss.append((y - y_hat) ** 2)
        parameters = update_parameters(parameters,y,y_hat,A1,X)
    print("Epoch -", i+1, "Loss", np.array(Loss).mean())
parameters
```
